```python
import ipywidgets as widgets
```

```python
c = widgets.Dropdown(
    options=gen,
    value='Music',
    description='Number:',
    disabled=False,
)
def graph(a):
  new = video[video['Gener']==a]
  channel=new.groupby(by = 'channelTitle')
  fin = channel.sum().sort_values(by = 'view_count',ascending = False).head()
  sns.barplot(x = fin.index , y = fin['view_count'])
  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python
c = widgets.Dropdown(
    options=gen,
    value='Music',
    description='Number:',
    disabled=False,
)
def graph(a):
  new = video[video['Gener']==a]
  channel=new.groupby(by = 'dislikes')
  fin = channel.sum().sort_values(by = 'view_count',ascending = False).head()
  sns.barplot(x = fin.index , y = fin['view_count'])
  plt.show()
```

```python
widgets.interact(graph,a=c)
```
